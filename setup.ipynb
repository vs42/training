### Windows

1.  Установка Python:
    *   Перейдите на официальный сайт python.org.
    *   Скачайте последнюю версию установщика.
    *   Запустите установщик и обязательно поставьте галочку Add Python to PATH. Это самый важный шаг.
    *   Нажмите Install Now и следуйте инструкциям.

2.  Установка VS Code:
    *   Перейдите на сайт code.visualstudio.com.
    *   Скачайте и запустите установщик для Windows.
    *   Следуйте инструкциям (можно просто нажимать "Далее").

---

### macOS

1.  Установка Python (через Homebrew):
    *   Сначала установите менеджер пакетов Homebrew. Откройте Терминал и выполните команду:
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
*   После установки Homebrew установите Python командой:
brew install python3
2.  Установка VS Code:
    *   Перейдите на сайт code.visualstudio.com.
    *   Скачайте .dmg файл (для Apple Silicon или Intel в зависимости от вашего Mac).
    *   Откройте скачанный файл и перетащите Visual Studio Code в папку Applications.

---

### Linux (Debian/Ubuntu)

1.  Установка Python:
    *   Python 3 обычно предустановлен. Проверьте его наличие, открыв Терминал и выполнив команду:
python3 --version
*   Если он не установлен или вы хотите последнюю версию, выполните команды:
sudo apt update
      sudo apt install python3 python3-pip
2.  Установка VS Code:
    *   Перейдите на сайт code.visualstudio.com.
    *   Скачайте .deb пакет для Debian/Ubuntu.
    *   Установите его через менеджер приложений или командой в терминале:
sudo dpkg -i <путь_к_файлу>.deb
*   *Альтернатива (через Snap):* sudo snap install --classic code
