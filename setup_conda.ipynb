# Установка и запуск Conda

## Windows

1. Установка Conda:
   * Перейдите на сайт https://repo.anaconda.com/miniconda/Miniconda3-latest-Windows-x86_64.exe
   * Скачайте и запустите установщик.
   * Следуйте инструкциям установщика (можно просто нажимать "Далее").
   * Завершите установку.

2. Запуск Conda:
   * Найдите в меню "Пуск" программу Anaconda Prompt (Miniconda3) и запустите её.
   * Conda будет активирована автоматически. В начале строки вы увидите (base).

---

## macOS

1. Установка Conda:
   * Откройте Терминал и выполните команды:
     
     curl -O https://repo.anaconda.com/miniconda/Miniconda3-latest-MacOSX-arm64.sh
     bash Miniconda3-latest-MacOSX-arm64.sh
        * Следуйте инструкциям установщика (нажимайте Enter, затем введите yes).
   * После установки инициализируйте conda:
     
     conda init zsh
        * Закройте и откройте Терминал заново.

2. Запуск Conda:
   * После перезапуска терминала conda будет активирована автоматически. В начале строки вы увидите (base).

---

## Linux (Debian/Ubuntu)

1. Установка Conda:
   * Откройте Терминал и выполните команды:
     
     wget https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh
     bash Miniconda3-latest-Linux-x86_64.sh
        * Следуйте инструкциям установщика (нажимайте Enter, затем введите yes).
   * После установки инициализируйте conda:
     
     conda init bash
        * Закройте и откройте Терминал заново.

2. Запуск Conda:
   * После перезапуска терминала conda будет активирована автоматически. В начале строки вы увидите (base).

# Установка Git

## Windows

1. Установка Git:
   * Перейдите на сайт https://git-scm.com/download/win
   * Скачайте установщик Git для Windows.
   * Запустите установщик и следуйте инструкциям (можно просто нажимать "Далее" и оставлять настройки по умолчанию).
   * Завершите установку.

2. Проверка установки:
   * Откройте командную строку (CMD) или PowerShell.
   * Выполните команду:
git --version
* Вы должны увидеть версию Git, например: git version 2.43.0

---

## macOS

1. Установка Git (через Homebrew):
   * Откройте Терминал и выполните команду:
brew install git
* Если Homebrew не установлен, сначала установите его:
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
2. Проверка установки:
   * В Терминале выполните команду:
git --version
* Вы должны увидеть версию Git, например: git version 2.43.0

---

## Linux (Debian/Ubuntu)

1. Установка Git:
   * Откройте Терминал и выполните команды:
sudo apt update
     sudo apt install git
2. Проверка установки:
   * В Терминале выполните команду:
git --version
* Вы должны увидеть версию Git, например: git version 2.43.0
